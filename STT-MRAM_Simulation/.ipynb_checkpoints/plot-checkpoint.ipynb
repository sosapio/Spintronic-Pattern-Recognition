```python

#!/usr/bin/env python
# coding: utf-8
#
# Copyright (c) 2020-2021 Arm Ltd.
# All rights reserved.
#
# SPDX-License-Identifier: BSD-3-Clause
# The codes are origianlly built from Fernando Garc´ıa-Redondo et al. A Compact Model for
# Scalable MTJ Simulation. arXiv:2106.04976 [cs]. June
# 2021. DOI: 10.48550/arXiv.2106.04976
# The codes are adjusted by Minyeok Wi

"""s-LLGS basic examples"""

import sllgs_solver as sllg_solver


def current(t):
    """Current waveform."""
    return -35e-6


def do_test(seed_0=100):
    """Simple s-LLGS tests."""
    sllg_solver.eq_info()
    
    # No thermal, no fake_thermal, solved with scipy_ivp
    # llg_a = sllg_solver.LLG(do_fake_thermal=False,
    #                         do_thermal=False,
    #                         i_amp_fn=current,
    #                         shape_ani_demag_n = True,
    #                         do_a_exchange = True,
    #                         seed=seed_0)
    # data_a = llg_a.solve_and_plot(100e-9,
    #                               scipy_ivp=True,
    #                               solve_spherical=True,
    #                               solve_normalized=True,
    #                               rtol=1e-4,
    #                               atol=1e-9)
    # No thermal, EMULATED THERMAL, solved with scipy_ivp
    llg_b = sllg_solver.LLG(do_fake_thermal=True,
                            d_theta_fake_th=1/30,
                            do_thermal=False,
                            i_amp_fn=current,
                            shape_ani_demag_n = True,
                            do_a_exchange = True,
                            seed=seed_0)
    data_b = llg_b.solve_and_plot(100e-9,
                                  scipy_ivp=True,
                                  solve_spherical=True,
                                  solve_normalized=True,
                                  max_step=1e-11,
                                  rtol=1e-4,
                                  atol=1e-9)
    

    llg_c = sllg_solver.LLG(do_fake_thermal=False,
                            do_thermal=True,
                            i_amp_fn=current,
                            shape_ani_demag_n = True,
                            do_a_exchange = True,
                            seed=seed_0)
    data_c = llg_c.solve_and_plot(100e-9,
                                  scipy_ivp=False,
                                  solve_spherical=False,
                                  solve_normalized=True,
                                  max_step=1e-13,
                                  method='stratonovich_heun')

    print(f'data ready to inspect: {data_b}, {data_c}')


do_test()
```

```python
import matplotlib.pyplot as plt
from sllgs_solver_Mod_Verilog import cart_from_spherical_fn, LLG

def plot_z_component_for_currents(currents, final_t=100e-9):
    """
    Plot the z-component of magnetization for varying currents over time.

    Args:
        currents (list): List of current values to test (in Amperes).
        final_t (float): Final simulation time (in seconds).
    """
    seed_0 = 100  # Seed for reproducibility
    plt.figure(figsize=(10, 6))

    for current_value in currents:
        # Define the current function for the given value
        def current_fn(t):
            return current_value

        # Initialize the solver with the current function
        llg_instance = LLG(
            do_fake_thermal=False,
            do_thermal=True,
            i_amp_fn=current_fn,
            shape_ani_demag_n = True,
            do_a_exchange=True,
            seed=seed_0
        )

        # Solve the system
        solution = llg_instance.solve_ode(
            final_t=final_t,
            scipy_ivp=False,
            solve_spherical=False,
            solve_normalized=True,
            max_step=1e-13,
            method='stratonovich_heun'
        )

        # Extract the z-component of magnetization
        states = solution.y
        times = solution.t
        xyz = cart_from_spherical_fn(states.T)
        m_z = xyz[:, 2]  # z-component

        # Convert current to microamps for the legend
        current_microamps = current_value * 1e6

        # Plot the z-component over time
        plt.plot(times * 1e9, m_z, label=f"Current = {-current_microamps:.0f} µA")

    # Add labels, title, and legend
    plt.xlabel("Time (ns)")
    plt.ylabel("Magnetization (m_z)")
    plt.title("Z-Component of Magnetization for Varying Currents")
    plt.legend()
    plt.grid()
    plt.show()

# Example usage
currents = [-18e-06, -35e-6, -100e-6, -150e-6]  # Currents in Amperes
plot_z_component_for_currents(currents)
```
